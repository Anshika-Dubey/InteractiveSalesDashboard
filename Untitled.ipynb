```python
import pandas as pd
```

```python
df=pd.read_csv('dataset.csv')
```

```python
print(df.head())
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
df.drop(columns=['ADDRESSLINE2', 'STATE', 'TERRITORY'], inplace=True)
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
df.fillna({'POSTALCODE':'No Data'}, inplace=True)
```

```python
df.columns = df.columns.str.strip()
```

```python
df['COUNTRY'] = df['COUNTRY'].str.strip()
df['PRODUCTLINE'] = df['PRODUCTLINE'].str.strip()
```

```python
import seaborn as sns
sns.boxplot(x=df['SALES'])
```

```python
# Check for outliers in 'SALES' using IQR
Q1 = df['SALES'].quantile(0.25)
Q3 = df['SALES'].quantile(0.75)
IQR = Q3 - Q1

# Remove outliers beyond 1.5 * IQR
df = df[~((df['SALES'] < (Q1 - 1.5 * IQR)) | (df['SALES'] > (Q3 + 1.5 * IQR)))]
```

```python
df.loc[:, 'Profit'] = df['SALES'] - (df['MSRP'] * df['QUANTITYORDERED'])
```

```python
df.head()
```

```python
total_orders = df['ORDERDATE'].nunique()
top_selling_product_line = df.groupby('PRODUCTLINE')['SALES'].sum().idxmax()
top_country = df.groupby('COUNTRY')['SALES'].sum().idxmax()
print(f"Total Orders: {total_orders}")
print(f"Top-Selling Product Line: {top_selling_product_line}")
print(f"Top Country by Sales: {top_country}")
```

```python
#key metrics
total_sales = df['SALES'].sum()
total_profit = df['Profit'].sum()
total_products_sold = df['QUANTITYORDERED'].sum()

print(f"Total Sales: {total_sales}")
print(f"Total Profit: {total_profit}")
print(f"Total Products Sold: {total_products_sold}")
```

```python
product_sales = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=True)
print(product_sales)
```

```python
#horizontal bar chart
import matplotlib.pyplot as plt
colors = ['#1f77b4' if i < len(product_sales)-3 else '#ff7f0e' for i in range(len(product_sales))]
plt.figure(figsize=(10, 6))
product_sales.plot(kind='barh', color=colors, edgecolor='black')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.title('Total Sales by Product Line')
plt.show()
```

```python
daily_sales = df.groupby('ORDERDATE')['SALES'].sum()
print(daily_sales)
```

```python
#plot line chart
import seaborn as sns
df = df.copy()
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')
df['Month'] = df['ORDERDATE'].dt.to_period('M')
monthly_sales = df.groupby('Month')['SALES'].sum()
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales.index.astype(str), y=monthly_sales.values, marker='o', color='yellow')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Over Time', fontsize=16, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(True, linewidth=0.6)
plt.tight_layout()
plt.show()
```

```python
region_sales = df.groupby('COUNTRY')['SALES'].sum()
print(region_sales)
```

```python
top_countries = region_sales.sort_values(ascending=False).head(10)
others = region_sales.sort_values(ascending=False).iloc[10:].sum()
top_countries['Others'] = others
plt.figure(figsize=(8, 8))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Region-wise Sales Distribution (Top 10 Countries)', fontsize=16, fontweight='bold')
plt.show()
```

```python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
from IPython.display import Markdown as md
display(md(f"**Total Orders:** `{total_orders}`"))
display(md(f"**Top-Selling Product Line:** `{top_selling_product_line}`"))
display(md(f"**Top Country:** `{top_country}`"))
app = dash.Dash(__name__)
app.layout = html.Div([
    html.H1('Interactive Sales Dashboard', style={'textAlign': 'center'}),
    
    html.Div([
        dcc.Dropdown(
            id='product-dropdown',
            options=[{'label': product, 'value': product} for product in df['PRODUCTLINE'].dropna().unique()],
            placeholder="Select a Product Line",
            style={'width': '50%'}
        ),
        dcc.Dropdown(
             id='country-dropdown',
            options= [{'label': country, 'value': country} for country in df['COUNTRY'].dropna().unique()],
            placeholder="Select a Country",
            style={'width': '50%'}
        )
    ]),

    dcc.Graph(id='sales-bar-chart')
])
@app.callback(
     Output('sales-bar-chart', 'figure'),
    [Input('product-dropdown', 'value'),
     Input('country-dropdown', 'value')]
)
def update_bar_chart(product_line, selected_country):
    filtered_df = df.copy()
    if product_line:
        filtered_df = filtered_df[filtered_df['PRODUCTLINE'] == product_line]
    if selected_country:
        filtered_df = filtered_df[filtered_df['COUNTRY'] == selected_country]
    fig = px.bar(
    filtered_df,
    x='COUNTRY',
    y='SALES',
    color='COUNTRY',
    color_discrete_sequence=px.colors.qualitative.Bold,  # Use a list of colors
    title=f"Sales for {product_line if product_line else 'All Products'} in {selected_country if selected_country else 'All Countries'}"
)
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
